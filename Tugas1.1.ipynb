```python
# 1. Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 2. Generate random data
x = np.linspace(0, 10, 100)
y = np.sin(x)
```

```python
# 3. Create a plot
plt.plot(x, y)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Sinusoidal Function')
```

```python
# 4. Add grid lines
plt.grid(True)
```

```python
# 5. Show the plot
plt.show()
```

```python
# 6. Define a function
def square(x):
    return x ** 2
```

```python
# 7. Call the function
result = square(5)
print("The square of 5 is:", result)
```

```python
# 8. Create a list
my_list = [1, 2, 3, 4, 5]
```

```python
# 9. Iterate over the list
for num in my_list:
    print(num)
```

```python
# 10. Calculate the sum of the list
sum_list = sum(my_list)
print("The sum of the list is:", sum_list)
```
