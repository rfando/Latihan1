```python
# Mengimport Library
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Unggah file dataset dari perangkat lokal Anda
from google.colab import files
uploaded = files.upload()
```

```python
# Ambil nama file yang diunggah
file_names = list(uploaded.keys())
```

```python
# Baca dataset menggunakan pandas
for file_name in file_names:
    df = pd.read_csv(file_name)
    print(f"Dataset {file_name} berhasil diunggah. Jumlah baris dan kolom: {df.shape}")
```

```python
# Menampilkan informasi data
print("Info sebelum cleaning:")
print(df.info())
```

```python
# Menangani nilai yang hilang
df = df.dropna()  # Menghapus baris yang memiliki nilai yang hilang
```

```python
# Tampilkan beberapa baris pertama data setelah pembersihan
print("\nData Setelah Pembersihan:")
print(df.info())
```

```python
# Mengatasi duplikat
df = df.drop_duplicates()
```

```python
# Tampilkan beberapa baris pertama dari dataset setelah cleaning
df.head()
```

```python
# Menampilkan nama kolom sebelum merapikan
print("Nama kolom sebelum merapikan:")
print(df.columns)
```

```python
# Merapikan nama kolom
df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_')
```

```python
# Menampilkan nama kolom setelah merapikan
print("\nNama kolom setelah merapikan:")
print(df.columns)
```

```python
# Menampilkan informasi data setelah merapikan
print("\nInfo setelah merapikan:")
print(df.info())
```

```python
# Tampilkan beberapa baris pertama dari dataset setelah merapikan
df.head()
```

```python
# Menyimpan dataset yang telah dibersihkan dan dirapikan ke file CSV
cleaned_file_name = 'dataset_cleaned.csv'
df.to_csv(cleaned_file_name, index=False)
```

```python
# Save the DataFrame to a CSV file
df.to_csv('data101.csv', index=False)
```
